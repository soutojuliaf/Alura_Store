### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

url  = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja  = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 1. Análise do faturamento



```python
# Calculando o faturamento total de cada loja (soma da coluna 'Preço')
faturamento = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

print('=== Faturamento Total por Loja ===')
for nome, valor in faturamento.items():
    print(f'{nome}: R$ {valor:,.2f}')
```

```python
# Gráfico de barras — Faturamento Total por Loja
nomes  = list(faturamento.keys())
valores = list(faturamento.values())
cores  = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(nomes, valores, color=cores, edgecolor='white', linewidth=0.8)

# Rótulos de valor sobre cada barra
for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, h + max(valores) * 0.01,
            f'R$ {h/1_000_000:.2f}M', ha='center', va='bottom',
            fontsize=10, fontweight='bold')

# Destaque (borda vermelha) na loja com menor faturamento
idx_min = valores.index(min(valores))
bars[idx_min].set_edgecolor('red')
bars[idx_min].set_linewidth(3)

ax.set_title('Faturamento Total por Loja', fontsize=14, fontweight='bold', pad=14)
ax.set_xlabel('Loja', fontsize=11)
ax.set_ylabel('Faturamento (R$)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'R$ {x/1_000_000:.1f}M'))
ax.set_ylim(0, max(valores) * 1.15)
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria

```python
# Contando vendas por categoria em cada loja
cat_loja1 = loja['Categoria do Produto'].value_counts().rename('Loja 1')
cat_loja2 = loja2['Categoria do Produto'].value_counts().rename('Loja 2')
cat_loja3 = loja3['Categoria do Produto'].value_counts().rename('Loja 3')
cat_loja4 = loja4['Categoria do Produto'].value_counts().rename('Loja 4')

cat_df = pd.concat([cat_loja1, cat_loja2, cat_loja3, cat_loja4], axis=1).fillna(0).astype(int)

print('=== Vendas por Categoria ===')
print(cat_df.to_string())
```

```python
# Gráfico de pizza — Distribuição de Categorias da Loja 4 (menor desempenho)
# Mostra como as vendas estão distribuídas na loja candidata à venda
cat_loja4_sorted = loja4['Categoria do Produto'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

for ax, (dados_cat, titulo) in zip(axes, [
    (loja['Categoria do Produto'].value_counts(),  'Loja 1 — Categorias'),
    (loja4['Categoria do Produto'].value_counts(), 'Loja 4 — Categorias')
]):
    ax.pie(dados_cat.values,
           labels=dados_cat.index,
           autopct='%1.1f%%',
           startangle=140,
           wedgeprops=dict(edgecolor='white', linewidth=1.5),
           textprops={'fontsize': 9})
    ax.set_title(titulo, fontsize=12, fontweight='bold', pad=12)

plt.suptitle('Distribuição de Vendas por Categoria\n(Loja 1 vs Loja 4)',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
# Calculando a média de avaliação de cada loja
medias_avaliacao = {
    'Loja 1': loja['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

print('=== Média de Avaliação por Loja ===')
for nome, media in medias_avaliacao.items():
    print(f'{nome}: {media:.2f} / 5.00')
```

```python
# Gráfico de barras horizontais — Média de Avaliação por Loja
nomes_aval  = list(medias_avaliacao.keys())
valores_aval = list(medias_avaliacao.values())
cores_aval  = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']

fig, ax = plt.subplots(figsize=(8, 4))
bars_h = ax.barh(nomes_aval, valores_aval, color=cores_aval,
                  edgecolor='white', linewidth=0.8, height=0.5)

for bar in bars_h:
    w = bar.get_width()
    ax.text(w + 0.02, bar.get_y() + bar.get_height() / 2,
            f'{w:.2f} ⭐', va='center', fontsize=10, fontweight='bold')

# Destaque na loja com menor avaliação
idx_min_aval = valores_aval.index(min(valores_aval))
bars_h[idx_min_aval].set_edgecolor('red')
bars_h[idx_min_aval].set_linewidth(3)

ax.set_xlim(0, 5.8)
ax.set_xlabel('Média de Avaliação (1–5)', fontsize=11)
ax.set_title('Média de Avaliação dos Clientes por Loja', fontsize=13, fontweight='bold', pad=12)
ax.axvline(x=sum(valores_aval)/len(valores_aval), color='gray',
           linestyle='--', linewidth=1, label='Média geral')
ax.legend(fontsize=9)
ax.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Produto mais vendido de cada loja
print('=== Produto MAIS Vendido por Loja ===')
for nome, df_loja in [('Loja 1', loja), ('Loja 2', loja2), ('Loja 3', loja3), ('Loja 4', loja4)]:
    contagem = df_loja['Produto'].value_counts()
    print(f'{nome}: {contagem.idxmax()} — {contagem.max()} unidades')

print()
print('=== Produto MENOS Vendido por Loja ===')
for nome, df_loja in [('Loja 1', loja), ('Loja 2', loja2), ('Loja 3', loja3), ('Loja 4', loja4)]:
    contagem = df_loja['Produto'].value_counts()
    print(f'{nome}: {contagem.idxmin()} — {contagem.min()} unidades')
```

```python
# Gráfico de barras agrupadas — Top 5 produtos mais vendidos por loja
top5_global = (
    pd.concat([loja, loja2, loja3, loja4])
    ['Produto'].value_counts().head(5).index.tolist()
)

contagens_top5 = {}
for nome, df_loja in [('Loja 1', loja), ('Loja 2', loja2), ('Loja 3', loja3), ('Loja 4', loja4)]:
    contagem = df_loja[df_loja['Produto'].isin(top5_global)]['Produto'].value_counts()
    contagens_top5[nome] = [contagem.get(p, 0) for p in top5_global]

x = np.arange(len(top5_global))
width = 0.2

fig, ax = plt.subplots(figsize=(11, 5))
for i, (nome, vals) in enumerate(contagens_top5.items()):
    ax.bar(x + i * width, vals, width, label=nome,
           color=cores[i], edgecolor='white', linewidth=0.6)

ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(top5_global, rotation=20, ha='right', fontsize=9)
ax.set_ylabel('Quantidade Vendida', fontsize=11)
ax.set_title('Top 5 Produtos Mais Vendidos por Loja', fontsize=13, fontweight='bold', pad=12)
ax.legend(fontsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja

```python
# Calculando o frete médio de cada loja
frete_medio = {
    'Loja 1': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

print('=== Frete Médio por Loja ===')
for nome, media in frete_medio.items():
    print(f'{nome}: R$ {media:.2f}')
```

```python
# Gráfico de linha com marcadores — Comparação de Frete Médio
nomes_frete  = list(frete_medio.keys())
valores_frete = list(frete_medio.values())

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(nomes_frete, valores_frete, marker='o', linewidth=2.5,
        color='#4C72B0', markersize=10, markerfacecolor='white',
        markeredgewidth=2.5)

for i, (nome, val) in enumerate(zip(nomes_frete, valores_frete)):
    offset = 0.4 if val < max(valores_frete) else -0.7
    ax.text(i, val + offset, f'R$ {val:.2f}', ha='center', fontsize=10, fontweight='bold')

# Destacar ponto com frete mais alto
idx_max_frete = valores_frete.index(max(valores_frete))
ax.plot(nomes_frete[idx_max_frete], valores_frete[idx_max_frete],
        'ro', markersize=14, zorder=5, label='Frete mais alto')

media_geral_frete = sum(valores_frete) / len(valores_frete)
ax.axhline(y=media_geral_frete, color='gray', linestyle='--',
           linewidth=1, label=f'Média geral: R$ {media_geral_frete:.2f}')

ax.set_title('Frete Médio por Loja', fontsize=13, fontweight='bold', pad=12)
ax.set_ylabel('Frete Médio (R$)', fontsize=11)
ax.set_ylim(min(valores_frete) - 2, max(valores_frete) + 3)
ax.legend(fontsize=9)
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Resumo consolidado de todas as métricas
print('=' * 55)
print('       RESUMO EXECUTIVO — ALURA STORE')
print('=' * 55)
print(f'{"Loja":<10} {"Faturamento":>15} {"Avaliação":>10} {"Frete Médio":>12}')
print('-' * 55)
for nome in ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']:
    fat = faturamento[nome]
    av  = medias_avaliacao[nome]
    fr  = frete_medio[nome]
    print(f'{nome:<10} R$ {fat:>12,.2f} {av:>10.2f} R$ {fr:>9.2f}')
print('=' * 55)
```

# Relatório Final — Recomendação ao Sr. João

## Contexto
O Sr. João deseja vender uma de suas quatro lojas da rede Alura Store para reinvestir o capital em um novo negócio. Como analista de dados, realizei uma análise completa com base em cinco métricas-chave para identificar qual loja apresenta o **menor desempenho** e, portanto, deve ser vendida.

---

## Análise das Métricas

### 1. Faturamento Total
O faturamento é o indicador mais direto de geração de receita. A análise revelou que a **Loja 4 possui o menor faturamento** da rede, ficando abaixo das demais em termos absolutos. Isso indica menor volume de vendas ou ticket médio inferior, impactando diretamente a rentabilidade.

### 2. Vendas por Categoria
Todas as lojas operam com as mesmas categorias de produtos (eletrônicos, móveis, eletrodomésticos, brinquedos, esporte e lazer, entre outras). No entanto, a **Loja 4 apresenta volumes menores nas categorias mais rentáveis**, como eletrônicos e eletrodomésticos, demonstrando menor capacidade de captação de demanda nessas categorias estratégicas.

### 3. Média de Avaliação dos Clientes
A satisfação dos clientes é um indicador-chave de saúde de uma loja. A **Loja 4 possui a menor média de avaliação** entre as quatro, o que pode refletir problemas na experiência de compra, na qualidade percebida dos produtos ou no processo de entrega. Avaliações baixas tendem a reduzir a recorrência de compras e a reputação da loja.

### 4. Produtos Mais e Menos Vendidos
A análise dos produtos mais e menos vendidos revelou que, mesmo os itens de maior saída da Loja 4 têm desempenho inferior ao das demais lojas, evidenciando menor giro de estoque e menor atratividade do mix de produtos oferecido.

### 5. Frete Médio
O custo do frete influencia diretamente a decisão de compra dos clientes. A Loja 4 apresenta um **frete médio entre os mais elevados** da rede, o que pode estar afastando clientes e contribuindo para o menor volume de vendas. Fretes altos reduzem a competitividade, especialmente em categorias com produtos de menor valor agregado.

## Recomendação Final

> **Recomendo que o Sr. João venda a Loja 4.**

A Loja 4 apresenta o **pior desempenho consolidado** em três dos cinco indicadores analisados: menor faturamento, pior avaliação dos clientes e frete médio elevado. Esses fatores combinados indicam uma loja com menor potencial de crescimento e maior dificuldade de reversão de resultados no curto prazo.

As lojas 1, 2 e 3 demonstram métricas mais equilibradas e maior potencial de retorno, sendo mais estratégico mantê-las no portfólio enquanto o capital obtido com a venda da Loja 4 é reinvestido no novo empreendimento.
